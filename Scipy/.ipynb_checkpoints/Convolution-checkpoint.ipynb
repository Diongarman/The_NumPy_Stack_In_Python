```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
g_im = Image.open('signal-2022-06-01-04-01-44-7792.jpg?width=985&height=801')
```

```python
six9ine_im = Image.open('ry9jkp_large.png')
```

```python
# resizing the second image to the same dimensions as the first one
six9ine_im = six9ine_im.resize(g_im.size)
```

```python
g_im = g_im.convert("RGBA")
six9ine_im = six9ine_im.convert("RGBA")
```

```python
alphaBlended1 = Image.blend(g_im, six9ine_im, alpha=.3)
```

```python
plt.imshow(alphaBlended1)
```

```python
for x in range(0,5):
    alphaBlended1 = Image.blend(g_im, six9ine_im, alpha=.1*x)
    plt.figure(figsize=(36, 32))
    plt.subplot(1,5,x)
    plt.imshow(alphaBlended1)
```
