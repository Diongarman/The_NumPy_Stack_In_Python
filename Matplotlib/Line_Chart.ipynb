```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,20,1000)

y = np.sin(x) + 0.2*x
```

```python
def mpl_line_plot(inpt,outpt, x_lab, y_lab, title):
    plt.plot(inpt, outpt)
    plt.xlabel(x_lab)
    plt.ylabel(y_lab)
    plt.title(title);
```

```python
mpl_line_plot(x, y, 'input', 'output', 'My Plot')

```
