```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_1 = np.linspace(0,2*np.pi,1000)
y_1 = np.linspace(0,2*np.pi,1000)

x_2 = np.linspace(0,2*np.pi,1000)
y_2 = np.linspace(0,2*np.pi,1000)
```

```python

plt.scatter(10*np.cos(x_1), 10*np.sin(y_1))
plt.scatter(5*np.cos(x_2), 5*np.sin(y_2))
```

```python
mu, sigma = 0, 1


x1_noise = np.random.normal(mu, sigma,1000)
y1_noise = np.random.normal(mu, sigma,1000)

x2_noise = np.random.normal(mu, sigma,1000)
y2_noise = np.random.normal(mu, sigma,1000)
```

```python
x1_noise
#x1_noise[x1_noise<-1].shape
```

```python
x1 = 10*np.cos(x_1) + 2.5*np.cos(x1_noise)
y1 = 10*np.sin(y_1) + 2.5*np.sin(y1_noise)

x2 = 5*np.cos(x_2) + 2*np.cos(x2_noise)
y2 = 5*np.sin(y_2) + 2*np.sin(y2_noise)
```

```python
plt.scatter(x1, y1)
plt.scatter(x2, y2)
```

```python
cat = np.zeros(2000)

cat[:1000] = 1
```

```python
#concat
X = np.concatenate((x1, x2), axis=0)
Y = np.concatenate((y1, y2), axis=0)
```

```python
X.shape
```

```python
Y.shape
```

```python
d = {'x': X, 'y': Y, 'target': cat}
df = pd.DataFrame(data=d)
```

```python
df.head()
```

```python
df.plot.scatter(x='x', y='y', c='target',colormap='viridis', colorbar=False)
```

```python
df['x^2'] = df['x'].apply(lambda x: x**2)
df['y^2'] = df['y'].apply(lambda y: y**2)
df['x*y'] = df.apply(lambda row: row['x']*row['y'],axis=1)
```

```python
df.head()
```

```python
df.to_csv("./output.csv", header=True, index=False)
```
